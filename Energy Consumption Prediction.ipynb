```python
df_energy['time'] = pd.to_datetime(df_energy['time'])

df_weather['dt_iso'] = pd.to_datetime(df_weather['dt_iso'])
```

```python
data = pd.merge(
   df_energy,
    df_weather,
    left_on='time',
    right_on='dt_iso',
    how='inner'
)
```

```python
# Drop unnecessary columns like weather icons or duplicate timestamps
data.drop(columns=['dt_iso', 'weather_icon', 'weather_id'], inplace=True, errors='ignore')
```

```python
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
```

```python
data.columns
```

```python
target = data['total load actual']
features = data.drop(columns=['time', 'total load actual', 'city_name', 'weather_main', 'weather_description'])
```

```python
scaler = MinMaxScaler()
features_scaled = scaler.fit_transform(features)
target_scaled = scaler.fit_transform(target.values.reshape(-1, 1))
```

```python


# Create sequences
def create_sequences(features, target, lookback):
    X, y = [], []
    for i in range(lookback, len(features)):
        X.append(features[i-lookback:i, :])
        y.append(target[i])
    return np.array(X), np.array(y)

lookback = 24  # Use 24 hours of data to predict the next step
X, y = create_sequences(features_scaled, target_scaled, lookback)

# Split into train/test sets
train_size = int(0.8 * len(X))
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Build LSTM Model
model = Sequential([
    LSTM(64, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])),
    Dropout(0.2),
    LSTM(32, return_sequences=False),
    Dropout(0.2),
    Dense(1)
])

model.compile(optimizer='adam', loss='mean_squared_error')
model.summary()

# Train the model
history = model.fit(X_train, y_train, epochs=12, batch_size=32, validation_data=(X_test, y_test), verbose=1)

# Evaluate
loss = model.evaluate(X_test, y_test)
print(f"Test Loss: {loss}")

# Predict and inverse scale
y_pred = model.predict(X_test)
y_pred_rescaled = scaler.inverse_transform(y_pred)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(range(len(y_test)), scaler.inverse_transform(y_test.reshape(-1, 1)), label='Actual',color="red")
plt.plot(range(len(y_pred)), y_pred_rescaled, label='Predicted',color="blue")
plt.legend()
plt.xlabel("Time")
plt.ylabel("Energy Consumption")
plt.title("Energy Consumption Prediction")
plt.show()
```

```python
import dash
import dash_core_components as dcc
import dash_html_components as html
import plotly.express as px
import pandas as pd

# Load the energy dataset
data = pd.read_csv('/content/energy_dataset.csv')

# Initialize the Dash app
app = dash.Dash(__name__)

# Create a line chart for the energy generation and consumption
def generate_line_chart(metric):
    fig = px.line(data, x='time', y=metric, title=f'{metric} Over Time', template='plotly_dark')
    return fig

# Layout of the dashboard
app.layout = html.Div([
    html.H1("Energy Consumption & Generation Dashboard", style={'textAlign': 'center', 'marginTop': '20px'}),

    html.Div([
        # Dropdown for energy metric
        html.Div([
            html.Label("Select Energy Metric"),
            dcc.Dropdown(
                id='energy-metric-dropdown',
                options=[
                    {'label': 'Total Load Actual', 'value': 'total load actual'},
                    {'label': 'Generation Solar', 'value': 'generation solar'},
                    {'label': 'Generation Wind Offshore', 'value': 'generation wind offshore'},
                    {'label': 'Generation Biomass', 'value': 'generation biomass'},
                    {'label': 'Generation Fossil Gas', 'value': 'generation fossil gas'}
                ],
                value='total load actual',
                style={'width': '100%'}
            ),
        ], style={'width': '48%', 'display': 'inline-block', 'padding': '10px'}),

        # Dropdown for time period
        html.Div([
            html.Label("Select Time Period"),
            dcc.Dropdown(
                id='time-period-dropdown',
                options=[
                    {'label': 'Hourly Data', 'value': 'hourly'},
                    {'label': 'Daily Data', 'value': 'daily'}
                ],
                value='hourly',
                style={'width': '100%'}
            ),
        ], style={'width': '48%', 'display': 'inline-block', 'padding': '10px'})
    ], style={'marginBottom': '20px'}),

    # Graph Section
    html.Div([
        dcc.Graph(id='energy-consumption-graph')
    ], style={'marginTop': '20px'})
])

# Callback function to update the graph
@app.callback(
    dash.dependencies.Output('energy-consumption-graph', 'figure'),
    [
        dash.dependencies.Input('energy-metric-dropdown', 'value'),
        dash.dependencies.Input('time-period-dropdown', 'value')
    ]
)
def update_graph(selected_metric, time_period):
    # Resample data for daily or hourly
    if time_period == 'daily':
        data['time'] = pd.to_datetime(data['time'])
        data_daily = data.resample('D', on='time').mean()  # Resample by day
    else:
        data_daily = data  # Use hourly data as-is

    # Generate the plot based on selected metric
    return generate_line_chart(selected_metric)

# Run the Dash app
app.run_server(debug=True, use_reloader=False, port=8050)
```
